### ***SECTION - 1 Read images***  

***In OpenCV we can Image read by use **imread()** function. There are methos we can read images***
<ul>
    <li> Gray Scale Mode [0] </li>
    <li> Colour Scale Mode [1] </li>
    <li> </li>
</ul>

```python
#Read image in Gray scale mode
image_1=cv2.imread('image-1.jpg',0)

#print the image
print(image_1)
```

```python
#Read image in Colour scale mode
image_2=cv2.imread('image-1.jpg',1)

#print the image
print(image_1)
```

```python
#Read image in Colour scale mode
image_3=cv2.imread('image-1.jpg',-1)

#print the image
print(image_1)
```

## ***Section - 2 Image Attribute And Disply image By Matplotlib****

***Printing image attribute like Datatype and Shape***

```python
#print Data type of image
print(image_1.dtype)
print(image_2.dtype)
print(image_3.dtype)

#print Shape of the image
print(image_1.shape)
print(image_2.shape)
print(image_3.shape)
```
